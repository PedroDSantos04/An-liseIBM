```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
df.columns
```

```python
df.info()
```

```python
df.head()
```

```python
plt.hist(df.Age)
plt.title("Distribuição de Idades")
```

```python
plt.hist(df.MonthlyIncome)
plt.title("Distruibuição de Salários")
```

```python
gender = df.Gender.value_counts(1)
fig, ax = plt.subplots()
ax.pie(gender, labels = gender.index, autopct='%1.1f%%')
plt.show()
```

```python
df.EducationField.value_counts().plot.bar()
plt.title('Campo Educacional dos Funcionários')
plt.show()
```

```python
sns.scatterplot(df, x= 'EducationField', y= "MonthlyIncome")
plt.show()
```

```python
sns.scatterplot(df, x= 'YearsAtCompany', y= "MonthlyIncome")
plt.show()
```

```python
corre = df.corr()
sns.heatmap(corre)
plt.show()
```
